## Data preparation Machine Learning

```python
df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

df_small = df.drop(['lunch'], axis=1)    #lunch is not relevant

o_encoder = OrdinalEncoder()

df_gender = df_small[['gender']]
df_small['gender'] = o_encoder.fit_transform(df_gender)
df_small['gender'] = df_small['gender'].astype(int)

df_race = df_small[['race/ethnicity']]
df_small['race/ethnicity'] = o_encoder.fit_transform(df_race)
df_small['race/ethnicity'] = df_small['race/ethnicity'].astype(int)

df_edu = df_small[['parental level of education']]
df_small['parental level of education'] = o_encoder.fit_transform(df_edu)
df_small['parental level of education'] = df_small['parental level of education'].astype(int)


df_prep = df_small[['test preparation course']]
df_small['test preparation course'] = o_encoder.fit_transform(df_prep)
df_small['test preparation course'] = df_small['test preparation course'].astype(int)

df_small.head()
```

## Split Data + Scaling

```python
from sklearn.model_selection import train_test_split

x = df_small.iloc[:, :-1]
y = df_small['total score']
#astype int?

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42, shuffle=True)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score

reg = LinearRegression().fit(x_train, y_train)

print("Unser Score Training Lineare Regression: ", reg.score(x_train, y_train))
print("Unser Score Test Lineare Regression: ", reg.score(x_test, y_test))

y_pre=reg.predict(x_test)


print("\nUnsere erwarteten Werte: ",y_pre)

reg.coef_

reg.intercept_
```
